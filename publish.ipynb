```python
collector = []
for source in sources:
    print(source)
    hold = holder.get(source)
    grill = hold.copy()
    grill = grill.rename(columns = {"company_":"company"})
    grill = grill.replace("\n",",", regex=True)
    grill = grill.drop_duplicates()
    grill = grill.reset_index(drop=False)
    grill['level_0'] = grill['level_0'].astype(str)
    grill['index'] = grill['index'].fillna(grill['Source']+grill['level_0'])
    
    del grill['level_0']
    collector.append(grill)
```

```python
import re
# Function to remove non-alphanumeric characters
def remove_special_characters(value):
    return re.sub(r'[^a-zA-Z0-9]', ' ', value)

# Applying the function to each value in the DataFrame
cols_to_clean = ['device',
'description',
'company',
'meta',
'manufacturer',
'manufacturer_country']

for column in cols_to_clean:
    mdr[column] = mdr[column].astype(str)
    mdr[column] = mdr[column].apply(remove_special_characters)
```

```python
from elasticsearch import Elasticsearch

# Configuration
es_host = "https://192.168.1.111:9248"
ca_cert_path = 'elastic-stack-ca.pem'  # Use the actual path to your exported CA certificate
basic_auth = ("username", "password")  # Replace with your actual credentials

# Initialize Elasticsearch client
es = Elasticsearch(
    es_host,
    verify_certs=True,
    ca_certs=ca_cert_path,
    basic_auth=basic_auth
)
```

```python
# Define index settings and mappings
index_settings = {
    "settings": {
        "analysis": {
            "analyzer": {
                "html_stripping_analyzer": {
                    "type": "custom",
                    "tokenizer": "standard",
                    "char_filter": ["html_strip"],
                    "filter": ["lowercase"]
                },
                "custom_html_strip_analyzer": {
                    "type": "custom",
                    "tokenizer": "standard",
                    "char_filter": ["html_strip"],
                },
                "standard_lowercase": {
                      "type": "custom",
                      "tokenizer": "standard",
                      "filter": ["lowercase"]
                }
            }
        }
    },
    "mappings": {
        "properties": {
            "index": {"type": "keyword"},
            "start_dates": {"type": "date", "format": "strict_date_optional_time||epoch_millis", "ignore_malformed": True},
            "end_dates": {"type": "date", "format": "strict_date_optional_time||epoch_millis", "ignore_malformed": True},
            "device": {"type": "text", "analyzer": "custom_html_strip_analyzer",  # For indexing: strips HTML, retains original case
        "fields": {
          "lowercase": {  ##// Sub-field for case-insensitive search
            "type": "text",
            "analyzer": "standard_lowercase"  ## // Custom analyzer to be defined in settings
          }
        }},
            "description": {"type": "text", "analyzer": "html_stripping_analyzer"},
            "company": {"type": "text", "analyzer": "html_stripping_analyzer"},
            "meta": {"type": "text", "analyzer": "html_stripping_analyzer"},
            "manufacturer": {"type": "text"},
            "manufacturer_country": {"type": "text"},
            "Source": {"type": "keyword"},
            "Updated": {"type": "date", "format": "strict_date_optional_time||epoch_millis"},
            "status": {"type": "text"}
        }
    }
}



# Specify the index name
index_name = "mdr"  # Replace with your actual index name

# Create the index with the defined settings and mappings
es.indices.create(index=index_name, body=index_settings)
```

```python
# Get all indices
indices = es.cat.indices(format='json')

# Print index names
for index in indices:
    print(index['index'])
```
