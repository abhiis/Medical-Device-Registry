```python
#path
import os
from glob import glob as g
pdf_files = g(os.path.join(os.getcwd(),"*.pdf"))
pdf_files = {i:file for i, file in enumerate(pdf_files)}
print("Found these files:", *(f"{key}: {file}" for key, file in pdf_files.items()), sep="\n")
```

```python
idx = 0
from pdfminer.high_level import extract_text

text = extract_text(pdf_files[idx])

# Do something with the extracted text
print(text)
```

```python
import ctypes
from ctypes.util import find_library
find_library("".join(("gsdll", str(ctypes.sizeof(ctypes.c_voidp) * 8), ".dll")))
```

```python
from tabula import read_pdf

# 'pages' can be set to 'all' to extract tables from all pages
dfs = read_pdf(pdf_files[idx], pages='all', multiple_tables=True)
collect_pdf_dataframe = []
# 'dfs' will be a list of DataFrames, each representing a table
for i, df in enumerate(dfs):
    print(f"Table {i}")
    collect_pdf_dataframe.append(df)
```

```python
for df in collect_pdf_dataframe:
    display(df)
```

```python
import pandas as pd
pd.concat(collect_pdf_dataframe)
```
