```python
from googletrans import Translator
translator = Translator()
from functools import lru_cache

@lru_cache(maxsize=None)
def translate_to_en(text):
    if pd.notna(text) and isinstance(text, str):
        translated_text = translator.translate(str(text), dest='en').text
        return translated_text
    else:
        return text
```

```python
kr_mdr_export['product name'] = kr_mdr_export['product name'].fillna("-")
```

```python
import re
df = kr_mdr_export
# Regular expression for matching Korean characters
korean_pattern = r'[\u1100-\u11FF\u3130-\u318F\uAC00-\uD7A3\uA960-\uA97F\uD7B0-\uD7FF]+'

# Function to check for the presence of Korean characters
def contains_korean(text):
    if pd.notna(text):
        if isinstance(text, str):
            if re.search(korean_pattern, text):
                return True
            return False

# Apply the function to the DataFrame column and get the rows with Korean characters
df['to_translate'] = df['product name'].map(lambda x: contains_korean(x))

df
```

```python
to_translate = df[df.to_translate == True]
translated = df[df.to_translate == False]
```
