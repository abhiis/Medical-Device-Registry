### For IVDR

```python
df_Eudamed = pd.DataFrame(columns = ['Actor_Organisation' ,'Actor_ID', 'Address' , 'Country' , 'Applicable_legislation',
                                    'UDI_Code' , 'Nomenclature', 'Basic_UDI', 'Risk_Class' ,'Device_name'])
# Actor_Organisation = []
# Actor_ID = []
# Address = []
# Country = []
# Applicable_legislation = []
# UDI_Code = []
# Nomenclature = []
# Basic_UDI = []
# Risk_Class = []
# Device_name = []

counter = 0
#Set the number of pages
while counter < 2:
    
    #time.sleep(2)
    for i in range(2, 25):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(4)
        button =driver.find_element(By.XPATH, "//tr['+str(i)+']/td/button").click()
        time.sleep(2)

        try:
            Actor_Organisation = driver.find_element(By.XPATH , "//div[@id='manufacturer']/div/dl/dd").text
            #Actor_Organisation.append(Actor_Org)
        except:
            #Actor_Organisation.append('n/a')
            Actor_Organisation = 'n/a'

        try:
            Actor_ID = driver.find_element(By.XPATH , "//div[@id='manufacturer']/div/dl[2]/dd").text
            #Actor_ID.append(Actors_ID)
        except:
            #Actors_ID.append('n/a')
            Actor_ID = 'n/a'

        try:
            Address = driver.find_element(By.XPATH , "//div[@id='manufacturer']/div/dl[3]/dd").text
            #Address.append(Addresss)
        except:
            #Address.append('n/a')
            Address = 'n/a'

        try:
            Country = driver.find_element(By.XPATH , "//div[@id='manufacturer']/div/dl[4]/dd").text
            #Country.append(Countrys)
        except:
            #Country.append('n/a')
            Country = 'n/a'
            
        items_list1=  driver.find_elements(By.XPATH , "//div[@id='basic-udi-data']/div[@class='container zebra']/dl[@class='row ng-star-inserted']")
        
        for i in items_list1:
            if i.text.startswith('Risk'):
                item = 'found'
                Risk_Class = i.text.replace('Risk class\n', "")

        if item == 'found':
            Risk_Class = Risk_Class
        else:
            Risk_Class = 'n/a'
            
        for i in items_list1:
            if i.text.startswith('Applicable'):
                item = 'found'
                Applicable_legislation = i.text.replace('Applicable legislation\n', "")

        if item == 'found':
            Applicable_legislation = Applicable_legislation
        else:
            Applicable_legislation = 'n/a'
        
            
        for i in items_list1:
            try:
                if i.text.startswith('Basic UDI-DI'):
                    item1 = 'found'
                    Basic_UDI = i.text.replace('Basic UDI-DI code\n', "")
            except:
                pass
        if item1 == 'found':
            Basic_UDI = Basic_UDI
        else:
            Basic_UDI = 'n/a'
            
        for i in items_list1:
            if i.text.startswith('Device name'):
                item = 'found'
                Device_name = i.text.replace('Device name\n', "")

        if item == 'found':
            Device_name = Device_name
        else:
            Device_name = 'n/a'
            
        items_list2=  driver.find_elements(By.XPATH , "//div[@id='udi-di-data']/div[@class='container zebra']/dl[@class='row ng-star-inserted']")
        
        for i in items_list2:
            if i.text.startswith('UDI-DI code'):
                item = 'found'
                UDI_Code = i.text.replace('UDI-DI code\n', "")

        if item == 'found':
            UDI_Code = UDI_Code
        else:
            UDI_Code = 'n/a'           
               
        for i in items_list2:
            if i.text.startswith('Nomenclature'):
                item = 'found'
                Nomenclature = i.text.replace('Nomenclature code(s)\n', "")

        if item == 'found':
            Nomenclature = Nomenclature
        else:
            Nomenclature = 'n/a'
            
        df_Eudamed=df_Eudamed.append({'Actor_Organisation': Actor_Organisation ,'Actor_ID': Actor_ID, 
                                      'Address' : Address , 'Country' : Country, 
                                      'Applicable_legislation' : Applicable_legislation,
                                    'UDI_Code' : UDI_Code , 'Nomenclature' : Nomenclature , 
                                      'Basic_UDI' : Basic_UDI, 'Risk_Class' : Risk_Class
                                      ,'Device_name' : Device_name},ignore_index=True)

        time.sleep(4)
        #driver.back()
        driver.find_element(By.XPATH, "//div[@class='d-print-none btn-back']/button").click()
        #time.sleep(1)
    
    #time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    Next = driver.find_element(By.XPATH , "//p-paginator[@class='ng-star-inserted']/div/a[@aria-label = 'Next page']").click()
    counter += 1
```
