```python
from bs4 import BeautifulSoup
import pandas as pd
def extract_page_data(response):
    soup = BeautifulSoup(response.text)
    table = soup.find('table', {'class':'oep-table'})
    hrefs = [tr.find_all('td')[-1].find('a')['href'] for tr in table.find_all('tr')[1:]] 
    page_data = pd.read_html(str(table))[0]
    page_data['href'] = hrefs
    return page_data
    
```

```python
extract_page_data(response)
```

```python
from translate import Translator

def translate_vietnamese_to_english(text):
    translator = Translator(to_lang="en")
    translation = translator.translate(text)
    return translation
```

```python
# Example usage
vietnamese_text = "Còn hiệu lực 23/12/2019"
english_translation = translate_vietnamese_to_english(vietnamese_text)
print("English Translation:", english_translation)
```
