```python
#path
import os
from glob import glob as g
pdf_files = g(os.path.join(os.getcwd(),"*.pdf"))
pdf_files = {i:file for i, file in enumerate(pdf_files)}
print("Found these files:", *(f"{key}: {file}" for key, file in pdf_files.items()), sep="\n")
```

```python
idx = 0
from pdfminer.high_level import extract_text

text = extract_text(pdf_files[idx])

# Do something with the extracted text
print(text)
```

```python
import camelot
# Explicitly set the path to the Ghostscript binary
#ghostscript.Ghostscript.binary = r'C:\\Users\\AbhishekSrivastava\\Anaconda3\\Library\\bin\\gsdll64.dll'  # Adjust path as needed

# Replace 'pdf_files[idx]' with your PDF file path
tables = camelot.read_pdf(pdf_files[idx], pages='all', flavor='stream')
collect_pdf_dataframe = []
# 'tables' is a list of Table objects, which you can iterate over and export to a DataFrame
for i, table in enumerate(tables):
    print(f"Table {i} on page {table.page}")
    df = table.df  # Convert to a DataFrame
    collect_pdf_dataframe.append(df)
```

```python
for df in collect_pdf_dataframe:
    display(df)
```
