```python
# Start actual data collection here
import os
import time
import requests
from requests.adapters import HTTPAdapter, Retry
from bs4 import BeautifulSoup
import pandas as pd
import warnings as ws
from datetime import datetime as date
import concurrent.futures
from tqdm import tqdm
import logging
logging.basicConfig(level=logging.DEBUG,  # Set the minimum level for messages to be captured
                    format='%(asctime)s - %(levelname)s - %(message)s',
                    filename='ANMAT_AR_spider.log')  # Log messages to a file
# Constants
BASE_URL = "http://helena.anmat.gob.ar/Boletin/"
MAX_RETRIES = 5
RETRY_INTERVAL = 1


headers = {
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
    'Accept-Language': 'en-US,en;q=0.9',
    'Cache-Control': 'max-age=0',
    'Connection': 'keep-alive',
    'Content-Type': 'application/x-www-form-urlencoded',
    'Origin': 'http://helena.anmat.gob.ar',
    'Referer': 'http://helena.anmat.gob.ar/Boletin/',
    'Upgrade-Insecure-Requests': '1',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
}

data = {
    '__EVENTTARGET': 'ctl00$ContentPlaceHolder1$gvTramites',
    '__EVENTARGUMENT': 'Page$1',
    '__LASTFOCUS': '',
    '__VIEWSTATE': '',
    '__VIEWSTATEGENERATOR': '87A27275',
    '__VIEWSTATEENCRYPTED': '',
    '__EVENTVALIDATION': '',
    'ctl00$ContentPlaceHolder1$ddEmpresasFiltro': '-1',
    'ctl00$ContentPlaceHolder1$txtPM': '',
    'ctl00$ContentPlaceHolder1$txtNombreDescriptivo': '',
    'ctl00$ContentPlaceHolder1$txtExpediente': '',
    'ctl00$ContentPlaceHolder1$txtMarca': '',
    'ctl00$ContentPlaceHolder1$txtModelo': '',
}

# To ensure consistent directory creation without errors
current_directory = os.getcwd()
output_directory = os.path.join(current_directory, "Fetched Data")

if not os.path.exists(output_directory):
    os.mkdir(output_directory)

ws.filterwarnings("ignore")

def clean_str(string):
    return string.replace('\xa0', " ").strip()

def create_retry_session():
    retry_strategy = Retry(
        total=MAX_RETRIES,
        backoff_factor=RETRY_INTERVAL,
        status_forcelist=[500, 502, 503, 504],  # Retry on these HTTP status codes
    )
    adapter = HTTPAdapter(max_retries=retry_strategy)
    session = requests.Session()
    session.mount("http://", adapter)
    session.mount("https://", adapter)
    return session

def fetch_data_with_retry(session, url, page_param):
    data['__EVENTARGUMENT'] = 'Page$'+str(page_param)
    retries = 0
    while retries < MAX_RETRIES:
        try:
            response = session.get(url, data=data, headers = headers, verify=False)
            response.raise_for_status()
            return response.text
        except requests.exceptions.RequestException as e:
            print(f"Error fetching data from {url}: {e}. Retrying in {RETRY_INTERVAL} seconds...")
            retries += 1
            time.sleep(RETRY_INTERVAL)

    print(f"Max retries reached. Unable to fetch data from {url}.")
    return None

def fetch_base_data(session, page_param):
    response_text = fetch_data_with_retry(session, BASE_URL, page_param)
    if not response_text:
        return None

    soup = BeautifulSoup(response_text, 'html.parser')
    table = soup.find('table')
    
    table = pd.read_html(str(table))[0]
    table['gatewayCaptureAt'] = date.today()
    logging.info("Page %d, Total results: %d", page_param, len(table))
   
    return table



def process_url(page_param, session):
    device_list = fetch_base_data(session, page_param)
    
    return device_list

def main():
    counter = 1
    page_params = list(range(1,3041))
    device_list_collector = pd.DataFrame()
    with create_retry_session() as session:
        with concurrent.futures.ThreadPoolExecutor(max_workers = 16) as executor:
            futures = [executor.submit(process_url, page_param, session) for page_param in page_params]

            for future in tqdm(concurrent.futures.as_completed(futures),
                               desc=f"Fetching data"):
                device_list = future.result()
                device_list_collector = pd.concat([device_list_collector, device_list])
                if len(device_list_collector) > (500*counter):
                    device_list_collector.to_excel(f'Fetched Data//chunks//ANMAT - AR - chunk {counter}.xlsx')
                    counter += 1
            
    print("Saving files")
    device_list_collector.to_excel(f'Fetched Data/ANMAT - AR - Device list_110124.xlsx')

if __name__ == '__main__':
    main()
```
