```python
# Start actual data collection here
import os
import time
import requests
from requests.adapters import HTTPAdapter, Retry
from bs4 import BeautifulSoup
import pandas as pd
import warnings as ws
from datetime import datetime as date
import concurrent.futures
from tqdm import tqdm

# Constants
BASE_URL = 'https://eservice.hsa.gov.sg/medics/md/mdEnquiry.do?action=getDeviceInfo&devId='
MAX_RETRIES = 5
RETRY_INTERVAL = 1

# To ensure consistent directory creation without errors
current_directory = os.getcwd()
output_directory = os.path.join(current_directory, "Fetched Data")

if not os.path.exists(output_directory):
    os.mkdir(output_directory)

ws.filterwarnings("ignore")

def clean_str(string):
    return string.replace('\xa0', " ").strip()

def create_retry_session():
    retry_strategy = Retry(
        total=MAX_RETRIES,
        backoff_factor=RETRY_INTERVAL,
        status_forcelist=[500, 502, 503, 504],  # Retry on these HTTP status codes
    )
    adapter = HTTPAdapter(max_retries=retry_strategy)
    session = requests.Session()
    session.mount("http://", adapter)
    session.mount("https://", adapter)
    return session

def fetch_data_with_retry(session, url):
    retries = 0
    while retries < MAX_RETRIES:
        try:
            response = session.get(url)
            response.raise_for_status()
            return response.text
        except requests.exceptions.RequestException as e:
            print(f"Error fetching data from {url}: {e}. Retrying in {RETRY_INTERVAL} seconds...")
            retries += 1
            time.sleep(RETRY_INTERVAL)

    print(f"Max retries reached. Unable to fetch data from {url}.")
    return None




def get_device_sg_mdr(session , device_id):
    create_url = BASE_URL+device_id
    response_text = fetch_data_with_retry(session, create_url)
    data = pd.DataFrame()
    if not response_text:
        return None

    soup = BeautifulSoup(response_text, 'html.parser')
    
    tables = soup.find_all('table', {'class':'fmTbl'})
    device_info = pd.read_html(str(tables[0]))[0]
    device_info = device_info.transpose()
    device_info.columns = device_info.iloc[0]
    device_info = device_info[1:]
    model_flag = False
    product_owner = ''
    registrant = ''
    for i, table in enumerate(tables[1:]):
        #print(table.text)
        if 'Product Owner' in table.text:
            product_owner = clean_str(tables[i+2].text)

        if 'Registrant' in table.text:
            registrant = clean_str(tables[i+2].text.strip())

        if 'Models' in table.text:
            model_flag = True
            models = table.find_next_sibling('table')
            models = pd.read_html(str(models))[0]
            models.columns = models.iloc[0]
            models = models[1:]
    if model_flag:
        models['Product Owner'] = product_owner
        device_info['Product Owner'] = product_owner
        device_info = pd.merge(device_info, models, how="outer")

    device_info['Registrant'] = registrant
    device_info['href'] = create_url

    return device_info
    


def process_url(guid, session):
    device_list = get_device_sg_mdr(session, guid)
    return device_list

def main():
    device_list_collector = pd.DataFrame()
    
    with create_retry_session() as session:
        with concurrent.futures.ThreadPoolExecutor(max_workers = 16) as executor:
            futures = [executor.submit(process_url, ids, session) for ids in holder]

            for future in tqdm(concurrent.futures.as_completed(futures), total=len(holder),
                               desc=f"Fetching data"):
                device_list = future.result()
                device_list_collector = pd.concat([device_list_collector, device_list])
            
    print("Saving files")
    device_list_collector.to_excel(f'Fetched Data/SG_MDR - Device list_300124.xlsx')
    

if __name__ == '__main__':
    main()
```
